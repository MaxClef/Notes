## Matplotlib example workflow

```python
# 1. Import matplotlib and ready it for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 3. Setup plot
fig, ax = plt.subplots(figsize=(10,5))    # figsize=(width, height)

# 4. Plot the data
ax.plot(x, y)

# 5. Costumize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 6. Save & Show (you save the whole Figure)
fig.savefig("saved-pics/sample-plot.png")
```




## Two options for Subplots



```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5));

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize =(10, 5))

#Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Pyplot vs Matplotlib (OO method)

* When plotting somthins quickly, it's ok to use pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
len(heart_disease), len(over_50)
```

```python
# Pyplot Method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(10, 6));
```

```python
# OO Method mixed with pyplot
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);

ax.set_xlim([45, 100]);  # Setting the x-axis limit from (45, 100)
```

```python
# OO Method from scratch 
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the Data
scatter = ax.scatter(x=over_50["age"], 
                     y=over_50["chol"],
                     c=over_50["target"])

# Costumize
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a Legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add o Horizontal line
ax.axhline(over_50["chol"].mean(),
            linestyle="--");
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10));
                                
                        ###########     ax0    #############
# Add data to "ax0"
scatter_0 = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Costumize "ax0"
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")

# Add a Legend
ax0.legend(*scatter_0.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--")

                        #############     ax1      ##########
 # Add data to "ax1"
scatter_1 = ax1.scatter(x=over_50["age"],
                        y=over_50["thalach"],
                        c=over_50["target"])

# Costumize data
ax1.set(title="Heart Disease and Maximum Heart Rate",
        xlabel="Age",
        ylabel="Maximum Heart Rate")

# Add Legend
ax1.legend(*scatter_1.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(y=over_50["thalach"].mean(),
             linestyle="--");

# Add title to the Figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use("seaborn-whitegrid")

# OO Method from scratch 
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the Data
scatter = ax.scatter(x=over_50["age"], 
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")          # cmap --> changes the colour scheme (can look it up online)

# Costumize
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a Legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add o Horizontal line
ax.axhline(over_50["chol"].mean(),
            linestyle="--");
```

```python
# Costumizing the "y" and "x" axis limitation



# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10));
                                
                        ###########     ax0    #############
# Add data to "ax0"
scatter_0 = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

# Costumize "ax0"
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")

# Change the x axis range
ax0.set_xlim([50, 80])

# Add a Legend
ax0.legend(*scatter_0.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--")

                        #############     ax1      ##########
 # Add data to "ax1"
scatter_1 = ax1.scatter(x=over_50["age"],
                        y=over_50["thalach"],
                        c=over_50["target"],
                        cmap="winter")

# Costumize data
ax1.set(title="Heart Disease and Maximum Heart Rate",
        xlabel="Age",
        ylabel="Maximum Heart Rate")

# Change the x and y axis range
ax1.set_xlim([50, 80])
ax1.set_ylim([60,200])
# Add Legend
ax1.legend(*scatter_1.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(y=over_50["thalach"].mean(),
             linestyle="--");

# Add title to the Figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
